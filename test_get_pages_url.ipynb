```python
import requests
import time


list_of_site_pages = []

# url of first page
url = 'https://www.imovirtual.com/comprar/apartamento/lisboa/?search%5Bregion_id%5D=11&search%5Bsubregion_id%5D=153&nrAdsPerPage=72'

for number in range(2,10000):
    new_url = url+'?page='+str(number)
    response = requests.get(new_url)  
    print(f'Response: {response}, page: {number}')
    if str(response) == '<Response [200]>':
        list_of_site_pages.append(new_url)
    else:
        break
list_of_site_pages.insert(0,url)

list_of_site_pages
```
