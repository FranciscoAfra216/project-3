```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
url = "https://www.imovirtual.com/comprar/lisboa/"
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content)
soup.prettify
```

```python
list_of_site_pages = []
for number in range(2,501):
    list_of_site_pages.append('https://www.imovirtual.com/en/comprar/apartamento/lisboa/'+'?page='+str(number))
list_of_site_pages.insert(0,'https://www.imovirtual.com/en/comprar/apartamento/lisboa/')
list_of_site_pages
```

```python
# select all listings on page, use class = col-md-content section-listing__row-content

listings = soup.find_all(attrs = {'class' : 'col-md-content section-listing__row-content'})

listings = soup.find_all(attrs = {'class' : 'offer-item-details'})

# extract URL to use as unique ID
#str(listings[0]).split("href")[1]
URLs = [house.a['href'] for house in listings]
URLs

# extract prices
prices = [int(house.find_all(attrs = {'class' : 'offer-item-price'})[0].text.replace(" ","").split("\n")[1].replace("€","")) for house in listings]
prices

# house_type 
house_type = [house.find_all(attrs = {'class': 'offer-item-rooms hidden-xs'})[0].text for house in listings]
house_type

# area
area = [float(house.find_all(attrs = {'class': 'hidden-xs offer-item-area'})[0].text.replace(" m²","").replace(",",".")) for house in listings]
area

# location
locations = [house.find_all(attrs = {'class': 'text-nowrap'})[1].text.split(" ")[-2].replace(",","") for house in listings]
locations

```

```python
URLs = []
prices = []
house_type = []
area = []
locations = []

for page in list_of_site_pages[:100]:
    print(page)
    response = requests.get(page)
    soup = BeautifulSoup(response.content)


    # select all listings on page, use class = col-md-content section-listing__row-content

    listings = soup.find_all(attrs = {'class' : 'col-md-content section-listing__row-content'})

    listings = soup.find_all(attrs = {'class' : 'offer-item-details'})

    URLs.append([house.a['href'] for house in listings])
    prices.append([house.find_all(attrs = {'class' : 'offer-item-price'})[0].text.replace(" ","").replace(",","").split("\n")[1].replace("€","") for house in listings])
    house_type.append([house.find_all(attrs = {'class': 'offer-item-rooms hidden-xs'})[0].text for house in listings])
    area.append([float(house.find_all(attrs = {'class': 'hidden-xs offer-item-area'})[0].text.replace(" m²","").replace(",",".").replace(" ","")) for house in listings])
    locations.append([house.find_all(attrs = {'class': 'text-nowrap'})[1].text.split(" ")[-2].replace(",","") for house in listings])
    time.sleep(0.001)
```

```python
master = pd.DataFrame()
master['location'] = [item for lister in locations for item in lister]
master['price'] = [item for lister in prices for item in lister]
master['tier'] = [item for lister in house_type for item in lister]
master['size'] = [item for lister in area for item in lister]
master['link'] = [item for lister in URLs for item in lister]

master.head(50)
```

```python
bin_names = ['very_small','small','medium','large','very_large']
master
master['size_bins'] = pd.cut(master['size'],5,labels = bin_names)
```

```python
master
grouped = master.groupby(['location','size_bins','tier']).agg({"price":'mean'}).reset_index()
grouped.head()
```

```python
def row_checker(row):
    for comb in grouped:
        if (comb['location'] == row['location']) and (comb['size_bins'] == row['size_bins']) and (comb['tier'] == row['tier']):
            return comb['price']
        else:
            pass
```

```python
new_df = pd.merge(master, grouped,  how='inner', left_on=['location','size_bins','tier'], right_on = ['location','size_bins','tier'])
```

```python
useful_df = new_df[['link','location','size_bins','tier','price_x','price_y']]
```

```python
useful_df.rename(columns = {'price_x':'price','price_y':'average_price'}, inplace = True)
useful_df.head()
```

```python
useful_df['profitability'] = useful_df['average_price']-useful_df['price']
useful_df = useful_df.sort_values(by = ['profitability'],ascending = False)
useful_df.head()
```
